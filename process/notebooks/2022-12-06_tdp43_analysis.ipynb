```python
import skimage
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pathlib
import tifffile
```

```python
data_dir = pathlib.Path('/lab/raw-data/EXP_JSM002_NGN2_DIV25_wholeplate')
mip_dir = pathlib.Path('/lab/processed-data/EXP_JSM002_NGN2_DIV25_wholeplate/max_projections')
all_prefixes = [x.name[:29] for x in data_dir.glob('*.tif')]
all_prefixes = np.sort(np.unique(all_prefixes))

channels = ['405', '488', '561', '647']

for prefix in all_prefixes:
    for ch in channels:

        z_section_files = [x.name for x in 
                           data_dir.glob(prefix + '*' + ch +'*')]
        z_section_files = np.sort(z_section_files)

        img_stack = np.zeros((z_section_files.shape[0], 1024, 1024))
        img_stack = img_stack.astype('uint16')

        for i_f, f in enumerate(z_section_files):
            tif = skimage.io.imread(data_dir / f)
            img_stack[i_f, :, :] = tif

        mip = img_stack.max(axis=0)
        # tifffile.imwrite(mip_dir / (prefix + ch + '.tif'), 
        #                  img_stack, metadata={'axes': 'YX'}, 
        #                  imagej=True, compression=None, dtype='uint16')
        skimage.io.imsave(mip_dir / (prefix + ch + '.tif'), mip)
```

```python
data_dir = pathlib.Path('/lab/raw-data/EXP_JSM002_NGN2_DIV25_wholeplate')
mip_dir = pathlib.Path('/lab/processed-data/EXP_JSM002_NGN2_DIV25_wholeplate/max_projections/merged')
all_prefixes = [x.name[:29] for x in data_dir.glob('*.tif')]
all_prefixes = np.sort(np.unique(all_prefixes))

channels = ['405', '488', '561', '647']

for prefix in all_prefixes:
    img_stack = np.zeros((z_section_files.shape[0], 4, 1024, 1024))
    img_stack = img_stack.astype('uint16')
    for i_ch, ch in enumerate(channels):

        z_section_files = [x.name for x in 
                           data_dir.glob(prefix + '*' + ch +'*')]
        z_section_files = np.sort(z_section_files)

        for i_f, f in enumerate(z_section_files):
            tif = skimage.io.imread(data_dir / f)
            img_stack[i_f, i_ch, :, :] = tif

    mip = img_stack.max(axis=0)
    tifffile.imwrite(mip_dir / (prefix[:-1] + '.tif'), 
                        mip, metadata={'axes': 'CYX'}, 
                        imagej=True, compression=None, dtype='uint16')
        # skimage.io.imsave(mip_dir / (prefix + ch + '.tif'), mip)
```
